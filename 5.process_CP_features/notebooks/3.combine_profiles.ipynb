# Normalize annotated single cells using negative control

## Import libraries

```python
import pathlib
import pprint

import pandas as pd
```

## Set paths and variables

```python
# directory where combined parquet file are located
data_dir = pathlib.Path("../data/2.sc_tracks_annotated_data/profiles")

# directory where the normalized parquet file is saved to
output_dir = pathlib.Path("../data/3.combined_data/")
output_dir.mkdir(exist_ok=True, parents=True)
```

```python
# get a list of files in the data directory
files = list(data_dir.glob("*.parquet"))
list_of_dfs = [pd.read_parquet(file) for file in files]
combined_df = pd.concat(list_of_dfs, ignore_index=True)
combined_df.to_parquet(output_dir / "combined_data.parquet")
print(combined_df.shape)
combined_df.head()
```
